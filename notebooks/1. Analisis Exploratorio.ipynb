```python
# 0. Se importan las librerías necesarias.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# 0.1. Se ajusta la configuración para mostrar todas las filas del DataFrame.
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# 1. Limpieza de datos.

# 1.1. Se realiza la carga de datos, y se imprimen los primeros 10 registros del DataFrame.
videoWatched_Dataframe = pd.read_csv('video_watched_raw_data.csv')
videoWatched_Dataframe.head(10)
```

```python
# 1.2. LA columna EVENT:_DATE se convierte a formato tipo fecha.
videoWatched_Dataframe['EVENT_DATE'] = pd.to_datetime(videoWatched_Dataframe['EVENT_DATE'])

# 1.3. Se verifican los tipos de datos de las columnas.
videoWatched_Dataframe.dtypes
```

```python
# 2. Consistencia de datos

# 2.1. Estadísticas descriptivas de las principales variables numéricas.
videoWatched_Dataframe[['EVENT_DATE', 'EVENT_HOUR', 'VIDEO_WATCHED_PERCENTAGE', 'VIDEO_LENGTH_MIN', 'VIDEO_RATING', 'USER_CLASS_PER_WEEK_GOAL']].describe()
```

```python
# 2.2.1. Se verifican el número de valores nulos dentro de las columnas del DataFrame.
videoWatched_Dataframe.isnull().sum().sort_values(ascending = False)[:29]
```

```python
# 2.2.2. Se verifican el porcentaje de valores nulos dentro de las columnas del DataFrame, sobre el total de registros.
(videoWatched_Dataframe.isnull().sum() / len(videoWatched_Dataframe) * 100).sort_values(ascending = False)[:29]
```

```python
#  2.3. Se imputan los valores nulos en las columnas de tipo texto del DataFrame.
excluirImputacion = ['EVENT_DATE', 'EVENT_HOUR', 'VIDEO_WATCHED_PERCENTAGE', 'VIDEO_LENGTH_MIN', 'VIDEO_RATING', 'USER_UD', 'USER_CLASS_PER_WEEK_GOAL']
videoWatched_Dataframe.loc[:, ~videoWatched_Dataframe.columns.isin(excluirImputacion)] = (videoWatched_Dataframe.loc[:, ~videoWatched_Dataframe.columns.isin(excluirImputacion)].fillna('Desconocido'))
videoWatched_Dataframe.head(10)
```

```python
# 3. Análisis de Datos.

# 3.1. Se calculan los principales KPIs de referencia respecto a los vídeos vistos de los últimos 90 días.

usuariosUnicos = videoWatched_Dataframe['USER_ID'].nunique(dropna = True)
numeroEventos = len(videoWatched_Dataframe)

reproduccionesPorUsuario = videoWatched_Dataframe.groupby('USER_ID', dropna = True).size()
media_ReproduccionesPorUsuario = float(reproduccionesPorUsuario.mean())
Percentil90_ReproduccionesPorUsuario = float(np.percentile(reproduccionesPorUsuario, 90))

media_ProcentajeVideoVisto = float(videoWatched_Dataframe['VIDEO_WATCHED_PERCENTAGE'].mean())

media_VideoRating = float(videoWatched_Dataframe['VIDEO_RATING'].mean())

usuariosViendoMas90P = videoWatched_Dataframe.loc[videoWatched_Dataframe['VIDEO_WATCHED_PERCENTAGE'] >= 0.9, 'USER_ID'].nunique() / usuariosUnicos
usuariosViendoMas80P = videoWatched_Dataframe.loc[videoWatched_Dataframe['VIDEO_WATCHED_PERCENTAGE'] >= 0.8, 'USER_ID'].nunique() / usuariosUnicos
usuariosViendoMas50P = videoWatched_Dataframe.loc[videoWatched_Dataframe['VIDEO_WATCHED_PERCENTAGE'] >= 0.5, 'USER_ID'].nunique() / usuariosUnicos

print('-' * 80)
print('Principales KPIs Vídeos Vistos - Últimos 90 Días'.center(80))
print('-' * 80, '\n')

#print(f'Rango de Fechas: {videoWatched_Dataframe['EVENT_DATE'].min().date()} / {videoWatched_Dataframe['EVENT_DATE'].max().date()}\n')
print(f"Rango de Fechas: {videoWatched_Dataframe['EVENT_DATE'].min().date()} / {videoWatched_Dataframe['EVENT_DATE'].max().date()}\n")

print(f'Número de Usuarios Únicos: {usuariosUnicos:,}')
print(f'Número de Eventos: {numeroEventos:,}\n')

print(f'Media de Reproducciones por Usuario: {media_ReproduccionesPorUsuario:,.2f}')
print(f'Percentil 90 de Reproducciones por Usuario: {Percentil90_ReproduccionesPorUsuario:,.0f}\n')

print(f'Media del Porcentaje de Vídeo Visto: {media_ProcentajeVideoVisto * 100:,.1f}%')
print(f'Media de la Valoración del Vídeo: {media_VideoRating * 100:,.1f}%\n')
    
print(f'Porcentaje de Usuarios que Vieron Más del 90% del Vídeo: {usuariosViendoMas90P * 100:,.1f}%')
print(f'Porcentaje de Usuarios que Vieron Más del 80% del Vídeo: {usuariosViendoMas80P * 100:,.1f}%')
print(f'Porcentaje de Usuarios que Vieron Más del 50% del Vídeo: {usuariosViendoMas50P * 100:,.1f}%\n')

print('-' * 80)
print('-' * 80)
```

```python

# 3.2. Se muestra la evolución histórica del número de usuarios, reproducciones y porcentaje promedio del vídeo visto en los últimos 90 días.
# Agrupar por fecha
dailyStats = (
    videoWatched_Dataframe.groupby('EVENT_DATE')
    .agg(
        TOTAL_USUARIOS = ('USER_ID', 'nunique'),
        TOTAL_REPRODUCCIONES = ('USER_ID', 'size'),
        AVG_WATCHED_PERCENTAGE = ('VIDEO_WATCHED_PERCENTAGE', 'mean')
    )
    .reset_index()
)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (12, 10))
# Usuarios y Reproducciones.
ax1.set_title('Evolución Diaria. Usuarios y Reproducciones (Últimos 90 Días).', fontsize = 12, fontweight = 'bold', pad = 10)
ax1.plot(dailyStats['EVENT_DATE'], dailyStats['TOTAL_USUARIOS'], linewidth = 1, color = '#1C4587', label = 'Usuarios Únicos')
ax1.set_xlabel('Fecha', fontsize = 9)
ax1.set_ylabel('Usuarios Únicos', fontsize = 9)
ax1.tick_params(labelsize = 9)
ax1b = ax1.twinx()
ax1b.plot(dailyStats['EVENT_DATE'], dailyStats['TOTAL_REPRODUCCIONES'], linewidth = 1, color = '#FF0000', alpha = 0.8, label = 'Reproducciones')
ax1b.set_ylabel('Reproducciones', fontsize = 9)
ax1b.tick_params(axis = 'y', labelsize = 9)
ax1.legend(loc = 'upper left', fontsize = 9)
ax1b.legend(loc = 'upper right', fontsize = 9)
# Promedio del Porcentaje de Video Visto.
ax2.set_title('Porcentaje Promedio de Video Visto por Día', fontsize = 12, fontweight = 'bold', pad = 10)
ax2.plot(dailyStats['EVENT_DATE'], dailyStats['AVG_WATCHED_PERCENTAGE'], linewidth = 1, color= '#34A853')
ax2.set_xlabel('Fecha', fontsize = 9)
ax2.tick_params(labelsize = 9)
plt.tight_layout(h_pad = 2)
plt.show()
```

```python
# 3.3. Se realiza un análisis de los patrones de reproducciones de videos por día y hora en Colombia.

ColombiaData = videoWatched_Dataframe[videoWatched_Dataframe['USER_COUNTRY'].str.contains('Colombia', case = False, na = False)].copy()
ColombiaData['DAY_OF_WEEK'] = ColombiaData['EVENT_DATE'].dt.day_name()

heatmapData = (
    ColombiaData.groupby(['DAY_OF_WEEK', 'EVENT_HOUR'])['USER_ID']
    .count()
    .reset_index()
    .pivot(index = 'DAY_OF_WEEK', columns = 'EVENT_HOUR', values = 'USER_ID')
)

orderedDays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmapData = heatmapData.reindex(orderedDays)

plt.figure(figsize=(12, 6))
ax = sns.heatmap(heatmapData, cmap = 'coolwarm', alpha = 0.8)

cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize = 9)

plt.title('Usuarios Únicos viendo Videos en Colombia (por día y hora)', fontsize = 12, fontweight = 'bold', pad = 10)
plt.xlabel('Hora del Día', fontsize = 9)
plt.ylabel('Día de la Semana', fontsize = 9)
plt.tight_layout()
plt.show()
```

```python
# 3.4. Número de usuarios en cada Bucket de Completion Video.

videoWatched_Dataframe['COMPLETION_BUCKET'] = pd.cut(videoWatched_Dataframe['VIDEO_WATCHED_PERCENTAGE'], bins = [0, 0.25, 0.50, 0.75, 1], labels = ['0 - 25%', '25 - 50%', '50 - 75%', '75 - 100%'], include_lowest = True, right = False)

completionSummary = (
    videoWatched_Dataframe.groupby('COMPLETION_BUCKET')
    .agg(
        TOTAL_EVENTOS = ('USER_ID', 'count'),
        USUARIOS_UNICOS = ('USER_ID', 'nunique')
    )
    .reset_index()
)

totalUsers = videoWatched_Dataframe['USER_ID'].nunique()
completionSummary['PORC_PARTICIPACION_USUARIOS'] = (completionSummary['USUARIOS_UNICOS'] / totalUsers * 100).round(1)
completionSummary
```

```python
# 3.5. Gráfico de Dispersión entre Duración y Porcentaje Visto de los Videos. 

plt.figure(figsize = (8, 6))
plt.title('Gráfico de Dispersión entre Duración y Porcentaje Visto de los Videos', fontsize = 12, fontweight = 'bold', pad = 10)

sns.scatterplot(data = videoWatched_Dataframe, x = 'VIDEO_LENGTH_MIN', y = 'VIDEO_WATCHED_PERCENTAGE', alpha = 0.6, s = 20, color = '#1c4587')

plt.xlabel('Duración (minutos)', fontsize = 9)
plt.ylabel('Porcentaje Visto', fontsize = 9)

plt.tick_params(labelsize = 9)

plt.tight_layout()
plt.show()
```

```python
# 3.6. Top 10 de los videos más Vistos en los últimos 90 Días.

videoSummary = videoWatched_Dataframe.groupby('VIDEO_NAME').agg(TOTAL_REPRODUCCIONES = ('USER_ID', 'count')).reset_index()
topVideos = videoSummary.sort_values(by = 'TOTAL_REPRODUCCIONES', ascending = False).head(10)

plt.figure(figsize = (10, 6))
plt.title('Top 10 Videos más Vistos', fontsize = 12, fontweight = 'bold', pad = 10)

ax = sns.barplot(data = topVideos, y = 'VIDEO_NAME', x = 'TOTAL_REPRODUCCIONES', palette = 'Blues_r')

for container in ax.containers:
    ax.bar_label(
        container,
        fmt = lambda x: f'{int(x):,}',
        label_type = 'edge',
        padding = 3,
        fontsize = 9,
        color = 'black'
    )

plt.xlabel('Número de Reproducciones', fontsize = 9)
plt.ylabel('')
plt.xlim(0, topVideos['TOTAL_REPRODUCCIONES'].max() * 1.1)

plt.tick_params(labelsize = 9)

plt.tight_layout()
plt.show()
```

```python
# 3.7. Top 10 de los videos más reproducidos en términos de duración en los últimos 90 Días.

videoWatched_Dataframe['ACTUAL_WATCHED_VIDEO_MIN'] = videoWatched_Dataframe['VIDEO_LENGTH_MIN'] * videoWatched_Dataframe['VIDEO_WATCHED_PERCENTAGE']
videoSummary = (
    videoWatched_Dataframe.groupby('VIDEO_NAME')
    .agg(TOTAL_ACTUAL_WATCHED_VIDEO_MIN = ('ACTUAL_WATCHED_VIDEO_MIN', 'sum'))
    .reset_index()
)

videoSummary = videoWatched_Dataframe.groupby('VIDEO_NAME').agg(ACTUAL_WATCHED_VIDEO_MIN = ('ACTUAL_WATCHED_VIDEO_MIN', 'sum')).reset_index()
topVideos = videoSummary.sort_values(by = 'ACTUAL_WATCHED_VIDEO_MIN', ascending = False).head(10)

plt.figure(figsize = (10, 6))
plt.title('Top 10 Videos con más Minutos Reproducidos', fontsize = 12, fontweight = 'bold', pad = 10)

ax = sns.barplot(data = topVideos, y = 'VIDEO_NAME', x = 'ACTUAL_WATCHED_VIDEO_MIN', palette = 'Blues_r')

for container in ax.containers:
    ax.bar_label(
        container,
        fmt = lambda x: f'{int(x):,}',
        label_type = 'edge',
        padding = 3,
        fontsize = 9,
        color = 'black'
    )

plt.xlabel('Tiempo de Reproducción (minutos)', fontsize = 9)
plt.ylabel('')
plt.xlim(0, topVideos['ACTUAL_WATCHED_VIDEO_MIN'].max() * 1.1)

plt.tick_params(labelsize = 9)

plt.tight_layout()
plt.show()
```
